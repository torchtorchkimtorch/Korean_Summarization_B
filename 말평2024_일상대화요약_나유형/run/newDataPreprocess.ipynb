```python
import json
```

```python
import os
directory_path = "../resource/data/new_datas/"
file_list = list(os.walk(directory_path))[0][2]
```

```python
def preprocess(file):
    template = {
        "id": "",
        "input": {
            "conversation": [
            ],
            "subject_keyword": []
        },
        "output": ""
    }
    prev_speaker = ""
    prev_utterance = ""
    template["id"] = file["id"]
    template['input']["subject_keyword"] = file['document'][0]['metadata']['topic'].split()[0]
    for i in range(len(file['document'][0]['utterance'])):
        curr_speaker = file['document'][0]['utterance'][i]['speaker_id']
        curr_utterance = file['document'][0]['utterance'][i]['form']
        if curr_utterance == "":
            pass
        else:
            if i==0:
                prev_speaker = curr_speaker
                prev_utterance = curr_utterance
            elif prev_speaker == curr_speaker:
                prev_utterance += f" {curr_utterance}"
                if i == len(file['document'][0]['utterance'])-1:
                    template["input"]["conversation"].append({'speaker':prev_speaker,"utterance":prev_utterance})
            else:
                if prev_utterance != "":
                    template["input"]["conversation"].append({'speaker':prev_speaker,"utterance":prev_utterance})
                prev_speaker = curr_speaker
                prev_utterance = curr_utterance
    template["output"] = file['document'][0]['SC']['main_summary']
    return template
```

```python
def json_len(template):
    total = 0
    for i in range(len(template['input']['conversation'])):
        total+= len(template['input']['conversation'][i]['utterance'])
    print(total)
```

```python
processed = []
for i in range(len(file_list)):
    file_path = f"../resource/data/new_datas/{file_list[i]}"
    with open(file_path,'r',encoding='utf-8') as f:
        file = json.load(f)
    processed_file = preprocess(file)
    json_len(processed_file)
    processed.append(processed_file)
```

```python
# 첫 번째 JSON 파일을 읽음
with open('../resource/data/일상대화요약_train.json', 'r', encoding='utf-8') as f1:
    data1 = json.load(f1)


# 두 개의 JSON 데이터를 병합
merged_data = data1 + processed

# 병합된 데이터를 새로운 JSON 파일에 저장
with open('../resource/data/new_train.json', 'w', encoding='utf-8') as f:
    json.dump(merged_data, f, ensure_ascii=False, indent=4)
```
